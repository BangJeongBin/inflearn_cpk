```python
# Q. 1부터 5까지 숫자를 한 줄씩 출력해보세요.
```

```python
# Q. 리스트에 들어있는 과일 이름을 하나씩 출력해보세요.
```

```python
fruits = ["사과", "바나나", "딸기"]

for fruit in fruits:
    print(fruit)
```

```python
# Q. 숫자 x가 5보다 작을 동안 계속 출력해보세요.
```

```python
x = 0

while x < 5:
    print(x)
    x += 1
```

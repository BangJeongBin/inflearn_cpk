```python
# Q. 숫자 x가 10보다 작거나 같으면 "10 이하입니다"를 출력하도록 코드를 작성해보세요.
```

```python
x = 7

if x <= 10:
    print("10 이하입니다")
```

```python
# Q. x가 10보다 크면 "10 초과", 그렇지 않으면 "10 이하"를 출력하도록 만들어보세요.
```

```python
x = 10

if x > 10:
    print("10 초과")
else:
    print("10 이하")
```

```python
# Q. `점수(score)`에 따라 결과를 다르게 출력해보세요.

# - 90점 이상이면 `"A"`
# - 80점 이상이면 `"B"`
# - 그 외는 `"C"`
```

```python
score = 85

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
else:
    print("C")
```
